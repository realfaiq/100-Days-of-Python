```python
#Chaining Convert the less space taken to int8 rather than 16 or 64
(autos
 [cols]
 .astype({'highway08': 'int8', 'city08': 'int16', 'comb08': 'int16'})
 .select_dtypes([int, 'int8'])
 .describe()
)
```

```python
#Chaining 
#Use integer to see all int-like columns
(autos
 [cols]
 .astype({'highway08': 'int8', 'city08': 'int16', 'comb08': 'int16', 'fuelCost08': 'int16',
         'range': 'int16', 'year': 'int16'})
 .select_dtypes('integer')
 .describe()
)
```

# Objects

```python
(autos
 [cols]
 .select_dtypes(object)
)
```

```python
(autos
 ['drive']
 .value_counts(dropna = False)
)
```

```python
#Where are the missing values
(autos
 [cols]
 .query('drive.isna()')
)
```

```python
#Drive and make (in.astype) to category
(autos
 [cols]
 .assign(cylinders = autos['cylinders'].fillna(0).astype('int8'),
        disp = autos.displ.fillna(0).astype('float16'),
        drive = autos.drive.fillna('Other').astype('category'))
 .astype({'highway08': 'int8', 'city08': 'int16', 'comb08': 'int16', 'fuelCost08': 'int16',
         'range': 'int16', 'year': 'int16', 'make': 'category'})
 .memory_usage(deep = True)
 .sum()
)
```

```python
#Trany Inspect
#Looks like two pieces of information embeded in column
(autos.trany.value_counts(dropna = False))
```

```python
#Add Automatic Speed for trany and then drop trany
(autos
 [cols]
 .assign(cylinders = autos['cylinders'].fillna(0).astype('int8'),
        disp = autos.displ.fillna(0).astype('float16'),
        drive = autos.drive.fillna('Other').astype('category'),
        automatic = autos.trany.str.contains('Auto'),
        speed = autos.trany.str.extract(r'(\d)+').fillna(20).astype('int8'))
 .astype({'highway08': 'int8', 'city08': 'int16', 'comb08': 'int16', 'fuelCost08': 'int16',
         'range': 'int16', 'year': 'int16', 'make': 'category'})
 .drop(columns = ['trany'])
 .memory_usage(deep = True)
 .sum()
)
```
