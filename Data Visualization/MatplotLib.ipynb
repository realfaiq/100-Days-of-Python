```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Basic Graph
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resize your Graph
plt.figure(figsize = (3,2), dpi = 100)

plt.plot(x,y, 'b^--', label = '2x')


#Short Hand
plt.plot(x,y, label = '2x', color = 'red', linewidth = 1, markersize = 12, markeredgecolor = 'blue')

#Line number 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label = 'x**2')
plt.plot(x2[5:], x2[5:]**2, 'r--', label = 'x**2')


plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()

#Save the graph
# plt.savefig('mygraph.png', dpi = 100)

plt.show()
```

### Loading Fifa Data

```python
fifa = pd.read_csv('C:/Users/Dell/Desktop/data-set/fifa_data.csv')
```

```python
fifa.head(10)
```

```python
#Histograms
bins = [40,50,60,70,80,90,100]
plt.hist(fifa['Overall'], bins = bins, color = 'lightblue')

plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distributions of Players Skills in Fifa 2018')

plt.xticks(bins)

plt.show()
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']

plt.pie([left, right], labels = labels, colors = ['red', 'purple'], autopct = '%.2f %%')

plt.title('Foot Preference of Fifa Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = [.4,.2,0,0,.4]
plt.title('Weight Distribution of Fifa Players (in LBS)')

plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)

plt.show()
```

```python
plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
labels = ['Barcelona', 'Real Madrid']

plt.title('Professional Football Team Comparsions')
plt.ylabel('FIFA Overall Rating')
boxes = plt.boxplot([barcelona, madrid], labels = labels)

for box in boxes['boxes']:
    box.set(color = 'red', linewidth = 2)

plt.show()
```
