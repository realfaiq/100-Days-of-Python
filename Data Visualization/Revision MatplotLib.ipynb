```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/Dell/Desktop/data-set/cereal.csv')
```

```python
data
```

```python
data.head(10)
```

```python
plt.figure(figsize = (20, 5))

plt.title('Different Cereals with amount of calories, protiens etc')

plt.plot(data['name'], data['protein'], 'r.-', label = 'Protein')
plt.plot(data['name'], data.sodium, 'b.-', label = 'Sodium')
plt.plot(data['name'], data.vitamins, 'g.-', label = 'Vitamins')

plt.xlabel('Serial Names')
plt.ylabel('Quantity')

plt.legend()
plt.show()
```

```python
bins = [40,50,60,70,80,90,100]
plt.hist(data['vitamins'], bins = bins, color = 'lightblue')

plt.xlabel('Vitamins')
plt.ylabel('Quantity')
plt.title('Distributions of Vitamins in different Cereals')

plt.xticks(bins)

plt.show()
```
