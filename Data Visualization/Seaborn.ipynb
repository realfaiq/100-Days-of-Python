### Bar Plot

```python
res = sns.barplot(x = mtcars['cyl'], y = mtcars['mpg'])
plt.show()
```

```python
res = sns.barplot(x = 'cyl', y = 'mpg', data = mtcars)
```

```python
res = sns.barplot(x = 'cyl', y = 'mpg', data = mtcars, color = 'red')
plt.show()
```

```python
res = sns.barplot(x = mtcars['am'], y = mtcars['mpg'], palette = 'Set2')
plt.show()
```

```python
res = sns.barplot(x = mtcars['gear'], y = mtcars['hp'], palette = 'rocket')
plt.show()
```

### CountPlot

```python
sns.countplot(x = 'cyl', data = mtcars, palette = 'Set1')
plt.show()
```

```python
sns.countplot(x = 'carb', data = mtcars, palette = 'Set2')
plt.show()
```

```python
sns.countplot(y = 'gear', data = mtcars, palette = 'rocket')
plt.show()
```

```python
sns.countplot(x = 'gear', hue = 'cyl', data = mtcars, palette = 'Set1')
plt.show()
```

### Dist Plot

```python
sns.distplot(mtcars.mpg, bins = 10, color = 'g')
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris)
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris, hue = 'species', size = 'species')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'species', palette = 'rocket')
```

### Lmplot

```python
sns.lmplot(x = 'petal_length', y = 'petal_width', data = iris)
```

```python
sns.lmplot(x = 'petal_length', y = 'petal_width', hue = 'species', data = iris, markers = ["o", "*", "^"])
```

### Box Plot

```python
sns.boxplot(x = 'species', y = 'sepal_width', data = iris)
plt.show()
```
