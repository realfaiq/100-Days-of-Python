### Bar Plot

```python
sns.barplot(x = cereal['protein'], y = cereal['calories'])
plt.show()
```

### Count Plot

```python
sns.countplot(x = 'protein', data = cereal, palette = 'Set1')
plt.show()
```

```python
sns.countplot(x = cereal['calories'], palette = 'Set2')
plt.show()
```

```python
sns.countplot(y = cereal['protein'], palette = 'Set1')
plt.show()
```

```python
sns.countplot(x = 'protein', hue = 'calories', data = cereal, palette = 'Set1')
plt.show()
```
