```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/Dell/Desktop/data-set/Parks.csv')
```

```python
df
```

```python
plt.figure(figsize = (8,5))
plt.title('Parks in Newyork')
plt.plot(df['site_name'], df['latitude'], 'b.-', label = 'Parks according to latitude')
plt.plot(df['site_name'], df['longitude'], 'r--', label = 'Parks according to longitude')
plt.xlabel('site_name')
plt.ylabel('latitude')
plt.legend()
plt.show()
```

```python
latitude = df['latitude'].count()
longitude = df['longitude'].count()
labels = ['Latitude', 'Longitude']
plt.pie([latitude, longitude], labels = labels, colors = ['red', 'purple'], autopct = '%.2f %%')
plt.title('Parks according to latitude and longitude')
plt.show()
```

```python
sns.countplot(x = df['latitude'], data = df, palette = 'Set2')
plt.show()
```

```python
sns.barplot(x = df['latitude'], y = df['longitude'])
plt.show()
```

```python
sns.scatterplot(x = df['latitude'], y = df['longitude'], data = df)
plt.show()
```
