```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
df['LoanAmount'].fillna(131, inplace = True)
df['Self_Employed'].fillna('No', inplace = True)
df['Credit_History'].fillna(0.0, inplace = True)
```

```python
self_employed_mapping = {'Yes': 1, 'No': 0}
df['self_employed_status'] = df['Self_Employed'].map(self_employed_mapping)
X = df[['self_employed_status', 'LoanAmount', 'Credit_History']]
loan_status_mapping = {'Y': 1, 'N': 0}
df['loan_status_results'] = df['Loan_Status'].map(loan_status_mapping)
y = df['loan_status_results']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
knn = KNeighborsClassifier(n_neighbors = 25, weights = 'uniform')
```

```python
knn.fit(X, y)
```

```python
prediction = knn.predict(X_test)
```
