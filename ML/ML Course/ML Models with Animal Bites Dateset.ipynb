# Data does not contain now any duplications or NAN values means we can proceed

```python
from sklearn.preprocessing import MinMaxScaler
import re
scaler = MinMaxScaler()
data['victim_zip'] = data['victim_zip'].apply(lambda x: re.sub("[^0-9]", "", str(x)))
data['victim_zip'] = data['victim_zip'].astype(float)
scaler.fit(data[['victim_zip']])
data['victim_zip_scaled'] = scaler.transform(data[['victim_zip']])
```

```python
data
```

```python
data['BreedIDDesc'].fillna('UNKNOWN', inplace = True)
```

```python
data['SpeciesIDDesc'].unique()
```

```python
data['SpeciesIDDesc'].fillna('OTHER', inplace = True)
```

```python
size_mapping = {'DOG': 1, 'CAT': 2, 'BAT': 3, 'RACCOON': 4, 'OTHER': 5, 'RABBIT': 6, 'HORSE': 7, 'SKUNK': 8, 'FERRET': 9}
data['SpeciesType'] = data['SpeciesIDDesc'].map(size_mapping)
```

```python
data
```

```python
gender_mapping = {'FEMALE': 1, 'MALE': 2, 'UNKNOWN': 0}
data['genders'] = data['GenderIDDesc'].map(gender_mapping)
```

```python
data
```

```python
data['WhereBittenIDDesc'].unique()
```

```python
bitten_area_mapping = {'BODY': 1, 'HEAD': 2, 'UNKNOWN': 0}
data['whereBitten'] = data['WhereBittenIDDesc'].map(bitten_area_mapping)
```

```python
data
```

```python
data['ResultsIDDesc'].unique()
```

```python
result_is_mapping = {'UNKNOWN': 0, 'POSITIVE': 1, 'NEGATIVE': 2, 'RESULTIDDESC': 3}
data['results'] = data['ResultsIDDesc'].map(result_is_mapping)
```

```python
data['results'].fillna(0, inplace = True)
```

```python
data
```

# Now that we got our features and labels right lets move to ML Models

```python
X = data[['SpeciesType', 'genders', 'whereBitten', 'victim_zip_scaled']]
y = data['results']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

# KNN

```python
from sklearn import neighbors, metrics
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors = 25, weights = 'uniform')
```

```python
knn.fit(X, y)
```

```python
prediction = knn.predict(X_test)
```

```python
print(prediction)
```

```python
accuracy = metrics.accuracy_score(y_test, prediction)
print("accuracy", accuracy)
```

# Linear Regression

```python
from sklearn import linear_model
```

```python
l_reg = linear_model.LinearRegression()
```

```python
model = l_reg.fit(X_train, y_train)
```

```python
l_predictions = model.predict(X_test)
```

```python
print(l_predictions)
```

```python
print("R^2", l_reg.score(X, y))
```
